```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('Russia social media.csv')

# Display the first 5 rows of the dataframexw
print(df.head(5))


# Save the top 10 rows of the dataframe to a CSV file
df.head(20).to_csv('top_20_rows.csv', index=False)

# To open and view the saved CSV file, you can read it back into a DataFrame
top_10_df = pd.read_csv('top_20_rows.csv')
print(top_10_df)
```

```python
root = "https://vl-nat-sec-hackathon-may-2024.s3.us-east-2.amazonaws.com/"

for rows in top_10_df.iterrows():
    urls = rows[1][-1].split(",")
    for url in urls:
        if ".mp4" in url:
            print(root + url)
        # from IPython.display import Image, display
        # display(Image(url=root + url))

```

```python
import requests

# Define the URL for the video to be downloaded
video_url = root + "media/vk/9043e959-0c9a-4507-bae3-c02439ceaaca.mp4"

# Send a GET request to download the video
response = requests.get(video_url)

# Check if the request was successful
if response.status_code == 200:
    # Write the video to a file
    with open("downloaded_video.mp4", "wb") as file:
        file.write(response.content)
    print("Video downloaded successfully.")
else:
    print("Failed to download video. Status code:", response.status_code)
```
